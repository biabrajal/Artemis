```python
import sqlite3
import pandas as pd
conn = sqlite3.connect('Loja.sqlite')
cursor = conn.cursor()
```

```python
query_clientes = """
    select
        c.Nome,
        p.PedidoID,
        p.DataPedido,
        p.ValorTotal
    from Clientes c
    left join Pedidos p
        on c.ClienteID = p.ClienteID;
"""

resposta = cursor.execute(query_clientes).fetchall()
resposta
df_clientes = pd.DataFrame(resposta, columns=["Nome","PedidoID", "DataPedido", "ValorTotal"])
df_clientes
```

```python
df_clientes = pd.read_sql_query(query_clientes, conn)
df_clientes
```

```python
query_clientes_tot = """
    select
        c.Nome,
        count(p.PedidoID) as QuantidadePedidos,
        sum(p.ValorTotal) as ValorTotalPedidos
    from Clientes c
    inner join Pedidos p
        on c.ClienteID = p.ClienteID
    group by c.Nome, c.ClienteID
    """
resposta = cursor.execute(query_clientes_tot).fetchall()
resposta
df_clientes_tot = pd.DataFrame(resposta, columns=["Nome", "QuantidadePedidos", "ValorTotalPedidos"])
df_clientes_tot
```

```python
df_clientes_tot = pd.read_sql_query(query_clientes_tot, conn)
df_clientes_tot
```

```python
query_pedidos = """
    select
        p.PedidoID,
        p.DataPedido,
        p.ValorTotal
    from Pedidos p
    left join Pagamentos pg
        on p.PedidoID = pg.PedidoID
    where pg.PagamentoID is null
    """
resposta = cursor.execute(query_pedidos).fetchall()
df_pedidos = pd.DataFrame(resposta, columns=["PedidoID", "DataPedido", "ValorTotal"])
df_pedidos
```

```python
df_pedidos = pd.read_sql_query(query_pedidos, conn)
df_pedidos
```
