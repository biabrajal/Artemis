```python
from datetime import datetime, timedelta

def parse_date(d):
    return datetime.strptime(d, "%Y-%m-%d")

def add_missing(lst):
    for i in lst:
        i[0] = parse_date(i[0])
        i.append("MISSING")

def match_valido(lst, int):
    if lst[1:] != int[1:]:
        return 0
    
    diff = (lst[0] - int[0]).days
    if (diff in (-1, 0, 1)):
        diff += 2
        return diff
    return 0
```

```python
def find_match(i, lst, indice, pontos):
    if indice==len(lst):
        return pontos
    int = lst[indice]
    if int[4]=="FOUND":
        return find_match(i, lst, indice+1, pontos)
    pontos[indice] = match_valido(i,int)
    return find_match(i, lst, indice+1, pontos)
```

```python
def reconcile_accounts (transactions1, transactions2):
    add_missing(transactions1)
    add_missing(transactions2)
    l = len(transactions2)
    for i in range (l):
        trans =transactions1[i]
        lst = find_match(trans, transactions2, 0, [0]*l)
        if any(x != 0 for x in lst):
            ind_max = lst.index(max(lst))
            transactions1[i][4] = "FOUND"
            transactions2[ind_max][4] = "FOUND"
    return transactions1, transactions2
```

```python
import csv
from pathlib import Path
from pprint import pprint
transactions1 = list(csv.reader(Path('transactions1.csv').open()))
transactions2 = list(csv.reader(Path('transactions2.csv').open()))
out1, out2 = reconcile_accounts(transactions1, transactions2)
```
